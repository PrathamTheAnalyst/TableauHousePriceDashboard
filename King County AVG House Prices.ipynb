# King County House Sales Dashboard in Tableau

This interactive dashboard shows information about average house prices, organized by a monthly timeline. It highlights how the monthly timeline affects the daily average house sales prices and displays the average prices in Washington as a heatmap. The dashboard provides insights into the distribution of house prices, the number of bedrooms and bathrooms, and the overall condition of houses for each day within the specified timeline. Users can adjust the displayed information by applying filters such as the year built, square footage of the living space, and the size of the lot in square feet.

Insights:
1. A sharp increase in the last week of May 2015 with almost 3 times the average price.
This insight indicates a significant anomaly or event that led to a substantial increase in house prices. 

2. The ideal number of bedrooms when searching for a house is 3 to 4, while the number of bathrooms is 2 to 3.
This insight provides valuable information for potential homebuyers, helping them understand the preferences in terms of the number of bedrooms and bathrooms that are considered ideal.

3. Average house prices are greater in the northern upper half of Washington than the southern part.
This geographical insight can influence decisions related to location and may have implications for real estate investment or homebuying strategies.

4. Overall, the distribution of average house prices is skewed to the right.
This general trend in housing prices gives a broad understanding of the market dynamics, indicating that there are more affordable homes and a decreasing number as prices rise.

5. The daily house prices between May 2014/15 show that most steep inclines in the average price are usually at the end of the month or the beginning of the month.
Understanding the timing of steep inclines in daily house prices can be valuable for both buyers and sellers, helping them plan transactions more strategically
